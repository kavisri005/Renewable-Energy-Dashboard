```python
#packages importing


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

print("Libraries loaded ✓")

```

```python
#load data

energy = pd.read_csv("energy_dataset.csv")
weather = pd.read_csv("weather_features.csv")

print("Energy Dataset Shape:", energy.shape)
print("Weather Dataset Shape:", weather.shape)
print("\nEnergy Columns:\n", energy.columns.tolist())
```

```python
#DateTime Conversion:

energy['time'] = pd.to_datetime(energy['time'], utc=True)
weather['dt_iso'] = pd.to_datetime(weather['dt_iso'], utc=True)

energy['year']    = energy['time'].dt.year
energy['month']   = energy['time'].dt.month
energy['hour']    = energy['time'].dt.hour
energy['weekday'] = energy['time'].dt.day_name()

print("DateTime converted ✓")
print("Date Range:", energy['time'].min(), "→", energy['time'].max())
```

```python
#Handle Missing Values

# Drop 100% null column
energy.drop(columns=['generation hydro pumped storage aggregated'], inplace=True)

# Fill numeric nulls with median
numeric_cols = energy.select_dtypes(include=np.number).columns
energy[numeric_cols] = energy[numeric_cols].fillna(energy[numeric_cols].median())

# Weather rain/snow fill with 0
weather[['rain_1h','rain_3h','snow_3h']] = weather[['rain_1h','rain_3h','snow_3h']].fillna(0)

print("Missing values handled ✓")
print("Nulls remaining - Energy:", energy.isnull().sum().sum())
print("Nulls remaining - Weather:", weather.isnull().sum().sum())
```

```python
# Check which column has nulls
print(energy.isnull().sum()[energy.isnull().sum() > 0])
```

```python
#Cell 5 cont - Drop null column:

energy.drop(columns=['forecast wind offshore eday ahead'], inplace=True)

print("Nulls remaining - Energy:", energy.isnull().sum().sum())
print("Energy shape:", energy.shape)
```

```python
# Remove duplicate rows from both datasets
energy.drop_duplicates(inplace=True)
weather.drop_duplicates(inplace=True)

print("Duplicates removed ✓")
print("Energy shape after cleaning:", energy.shape)
print("Weather shape after cleaning:", weather.shape)
```
